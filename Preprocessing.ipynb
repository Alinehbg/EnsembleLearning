```python
dataframe_path = '/content/drive/MyDrive/Ensemble project/data.xlsx'
```

```python
df = pd.read_excel(dataframe_path, index_col=0)
df.head()
```

```python
df.shape
```

```python
nan_count = df.isna().sum()
print(nan_count)
```

```python
# remove rows with missing values for the neighbourhood_group column
df.dropna(subset=['neighbourhood_group'], inplace=True)
```

```python
nan_count_checked = df.isna().sum()
print(nan_count_checked)
```

```python
df.groupby(['neighbourhood_group']).size()
```

```python
#replacing all NaN values in 'reviews_per_month' with 0
df.fillna({'reviews_per_month':0}, inplace=True)

#examing changes
df.reviews_per_month.isnull().sum()
```

```python
nan_count = df.isna().sum()
print(nan_count)
```

```python
# removing the columns that are :
#either non-interesting for our target: name,neighbourhood, host_name, latitude,	longitude,host_id
#I choose to remove last_review but let me know if you disagree
df=df.drop(['name','host_name', 'neighbourhood',	'latitude',	'longitude','last_review', 'host_id'], axis=1)
```

```python
df.head()
```

```python
# Performing one-hot encoding on the column 'neighbourhood_group' and 'room_type' because they contain categorical values
encoded_data = pd.get_dummies(df, columns=['neighbourhood_group','room_type'])

# Save the encoded data to a new CSV file
encoded_data.to_csv('data_processed.csv', index=False)
```
